<h3>
    <font color ='#4169E1'>
                            1.2 Taxa de Mortalidade Infantil<a id=1.2></a>
    </font>  
</h3>

<p>
    <font size=3>
        A taxa de <b>mortalidade infantil</b> é obtida por meio do número de crianças de um determinado local (cidade, região, país, continente) que morrem antes de completar 1 ano, a cada mil nascidas vivas.
    </font>  
</p>
<p>
    <font size=3>
        Esse dado é um aspecto  de fundalmental importância para avaliar a qualidade de vida, por meio dele, é possível obter informações sobre a eficácia dos serviços públicos, tais como: saneamento básico, sistema de saúde, disponibilidade de remédios e vacinas, acompanhamento médico, educação, maternidade, alimentação adequda, entre outros.
    </font>  
</p>
<p>
    <font size=3>
        Como é de se esperar as regiões mais atingidas pela <b>mortalidade infantil</b> são as com menor poder financeiro. A falta de assistência e de orientação às grávidas e a deficiência na assistência hospital aos recém-nascidos são um dos principais motivos.
    </font>  
</p>

<img src='https://upload.wikimedia.org/wikipedia/commons/thumb/9/93/Infant_Mortality_Rate_World_map.png/640px-Infant_Mortality_Rate_World_map.png'>

-------------------------

<h3>
    <font color ='#4169E1'>
                            1.3 Contexto<a id=1.3></a>
    </font>  
</h3>

<p>
    <font size=3>
        Reduzir a mortalidade infantil é um indicador chave para o progesso humano. A ONU espera que, até 2030, os países acabem com as mortes evitáveis de recém-nascidos e crianças menores de 5 anos, reduzindo a mortalidade para pelo menos 25 por 1.000 nascidos vivos.
    </font>  
</p>
<p>
    <font size=3>
        A <b>mortalidade materna</b>, obviamente, anda junto com a taxa de <b>mortalidade infantil</b>, a OMS calcula que cerca de 830 mulheres morrem todos os dias no mundo, devido a complicações na gravidez. A grande maioria dessas mortes (94%) ocorreu em ambientes de poucos recursos e poderia ter sido <u>evitada</u>.
    </font>  
</p>
<p>
    <font size=3>
       Diante desse problema, os Cardiotocogramas são uma opção simples e de baixo custo para avaliar a saúde fetal, permitindo que os profissionais de saúde atuem na prevenção da mortalidade infantil e materna. O equipamento funciona da seguinte forma, verifica os batimentos cardíacos e o bem estar do bebê, e é feito com sensores ligados à barriga da gestante que coletam estas informações.
    </font>  
</p>

<img src='https://miro.medium.com/max/558/1*bxVffEJZKnlJ5WfPMgefVg@2x.png' width=200 heigth=200>

```python
X=df.drop(["fetal_health"],axis=1)
y=df["fetal_health"]
col_names = list(X.columns)

scaler = StandardScaler()
X_pad = scaler.fit_transform(X)
X_pad = pd.DataFrame(X_pad, columns=col_names)
```

```python
plt.figure(figsize=(20,10))
sns.boxenplot(data = X_pad)
plt.xticks(rotation=90)
plt.show()
```

----------------------

<p>
    <font size=3>
        Dividindo o dataset padronizado em Treino e Teste para treinarmos os modelos e verificar a acurácia.
    </font>  
</p>

```python
X_train, X_test, y_train,y_test = train_test_split(X_pad, y, test_size=0.25, random_state=0)
```

<p>
    <font size=3>
        Construindo Pipelines para treinar e rodar os modelos.
    </font>  
</p>

```python
pipeline_lr = Pipeline([('lr_classifier', LogisticRegression(max_iter=200,random_state=0))])
pipeline_dt = Pipeline([('dt_classifier', DecisionTreeClassifier(random_state=0))])
pipeline_rf = Pipeline([('rf_classifier', RandomForestClassifier(random_state=0))])
pipeline_gb = Pipeline([('gb_classifier', GradientBoostingClassifier(random_state=0))])
pipeline_kn = Pipeline([('kn_classifier', KNeighborsClassifier())])


pipelines = [pipeline_lr, pipeline_dt, pipeline_rf, pipeline_gb, pipeline_kn]

pipe_dict = {0: 'Logistic Regression', 1: 'Decision Tree', 2: 'Random Forest', 3: 'Gradient Boosting', 4: 'KNeighbors'}

for pipe in pipelines:
    pipe.fit(X_train, y_train)

cv_results_accuracy = []
for i, model in enumerate(pipelines):
    cv_score = cross_val_score(model, X_train,y_train, cv=10 )
    cv_results_accuracy.append(cv_score)
    print("%s: %f " % (pipe_dict[i], cv_score.mean()))
```

<p>
    <font size=3>
        A melhor acurácia entre os modelos no dataset padronizado é do <i>Gradient Boosting</i>.
    </font>  
</p>

```python
pipeline_lr = Pipeline([('lr_classifier', LogisticRegression(max_iter=200,random_state=0))])
pipeline_dt = Pipeline([('dt_classifier', DecisionTreeClassifier(random_state=0))])
pipeline_rf = Pipeline([('rf_classifier', RandomForestClassifier(random_state=0))])
pipeline_gb = Pipeline([('gb_classifier', GradientBoostingClassifier(random_state=0))])
pipeline_kn = Pipeline([('kn_classifier', KNeighborsClassifier())])


pipelines = [pipeline_lr, pipeline_dt, pipeline_rf, pipeline_gb, pipeline_kn]

pipe_dict = {0: 'Logistic Regression', 1: 'Decision Tree', 2: 'Random Forest', 3: 'Gradient Boosting', 4: 'KNeighbors'}

for pipe in pipelines:
    pipe.fit(X_train, y_train)

cv_results_accuracy = []
for i, model in enumerate(pipelines):
    cv_score = cross_val_score(model, X_train,y_train, cv=10 )
    cv_results_accuracy.append(cv_score)
    print("%s: %f " % (pipe_dict[i], cv_score.mean()))
```

<h2>
    <font color ='#4169E1'>
                            5. Oversampling <a id=5></a>
    </font>  
</h2>

<p>
    <font size=3>
        <i>Oversampling</i>, ou sobamostragem, é uma forma de balancear os dados, podemos visualizar no gráfico da variável resposta, no 3.6.1, que existem muito mais informações sobre os dados que são normais, do que suspeitos e patológicos, podendo gerar um certo enviesamento dos dados. Com essa técnica, iremos aumentar o número de informações, para que fiquem balanceados, na tentativa de melhorarmos nosso modelo.
    </font>  
</p>

```python
#importando a bliblioteca
from imblearn.over_sampling import SMOTE
```

```python
#Aumentando as informações da nossa base de dados
smote = SMOTE()
X_over, y_over = smote.fit_sample(X,y)
```

```python
#verificando seu novo tamanho
X_over.shape, y_over.shape
```

```python
#separando os dados, em treino e teste
X_treino_o, X_teste_o, y_treino_o, y_teste_o = train_test_split(X_over, y_over, test_size=0.2, stratify=y_over)
```

```python
#selecionando melhores parâmetros
param_space = {
              "learning_rate": [0.05, 0.075, 0.1, 0.25, 0.5, 0.75, 1], 
              "n_estimators": [5,50, 200, 350, 500, 750],
              "max_depth": [1, 3, 5, 7, 9],
              "max_features":["log2","sqrt"]
              }

gbc=GradientBoostingClassifier()

gradient_boosting_search = RandomizedSearchCV(gbc, param_space,
                                             n_iter=32, scoring='accuracy', verbose=True, cv=5, n_jobs=-1, random_state=42)


gradient_boosting_search.fit(X_treino_o, y_treino_o)
```

```python
gradient_boosting_search.best_params_
```

```python
gradient_boosting_search.best_estimator_
```

```python
#rodando o modelo
model_gb = GradientBoostingClassifier(learning_rate=0.5,
                                  max_depth=9, max_features="sqrt",  
                                  n_estimators=350, random_state=42)

model_gb.fit(X_treino_o, y_treino_o)
                                      
print(f'Coeficiente de determinação da base de treino: {model_gb.score(X_treino_o, y_treino_o)}')
print(f'Coeficiente de determinação da base de teste: {model_gb.score(X_teste_o, y_teste_o)}')
```

```python
predict = model_gb.predict(X_teste_o)

accuracy = accuracy_score(y_teste_o, predict)
print(f"Acurácia do modelo na base de teste: {accuracy}")
```
